# Sub Assignment 1:Processing the Dataset

```python
#Data Exploration
#Display the first few rows of the dataset
print(df.head())
```

```python
#Get overall information of the Dataset:
df.info()
```

```python
#The Statistical information of the give data in dataset:
df.describe()
```

```python
#find null values if there:
df.isna()
```

```python
#Handle Missing Values:
df.fillna(df.mean(),inplace=True)
```

```python
# Since the feature 'State & District' is in the form of:
# 'district_code:[District_Code] state_name: [State_Name] district_name: [District_Name]'
# So, use str.split() method to extract District_Code, State_Name, District_Name From the string
# Method takes argument each 'State & District' values and
# Return list : [District_Code, State_Name, District_Name]
def separate_cols(row):
  divided = row.split(',')
  dist_code = int(divided[0].replace('district_code:', ''))
  state_name = divided[1].replace(' state_name:', '')
  dist_name = divided[2].replace(' district_name:', '')
  output = list((dist_code, state_name, dist_name))

  return output
```

```python
# Create New columns in the DataFrame for separated data
# Used Series.map() method and above created separate_cols() function

df['District_Code'] = df['State & District'].map(lambda x: separate_cols(x)[0])
df['State_Name'] = df['State & District'].map(lambda x: separate_cols(x)[1])
df['District_Name'] = df['State & District'].map(lambda x: separate_cols(x)[2])
```

```python
# Now safe to delete 'State & District' Feature
df.drop('State & District', axis=1, inplace=True)
```

#### 3.How much SC Male and Female Population present in each State :

```python
# Find sum of SC, Male_SC and Female_SC population in each state
# using groupby() method

state_SC_counts = df.groupby('State_Name')['SC'].sum()
state_Male_SC_counts = df.groupby('State_Name')['Male_SC'].sum()
state_Female_SC_counts = df.groupby('State_Name')['Female_SC'].sum()

# Create DataFrame from all these counts
lst = [state_SC_counts, state_Male_SC_counts, state_Female_SC_counts]
merged_SC_counts = pd.concat(lst, axis=1)
```

```python
# Create an array for the x-axis positions
x = np.arange(len(merged_SC_counts))

categories = merged_SC_counts.index.tolist()

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the simple bar
ax.bar(x-0.2, merged_SC_counts['SC'], label='SC Count', width=0.35)

# Plot the stacked bars
ax.bar(x+0.2, merged_SC_counts['Male_SC'], label='Male SC Counts', width=0.35)
ax.bar(x+0.2, merged_SC_counts['Female_SC'], label='Female SC Counts', width=0.35,
       bottom=merged_SC_counts['Male_SC'])

# Set labels and title
ax.set_xlabel('Names of States')
ax.set_ylabel('Population')
ax.set_title('Male and Female SC Population in each State')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
